## Exercise 3: Multivariate Linear Regression on sample data

> In most use cases, there will be more than one input variable to describe a desired target variable. Therefore, the third and the fourth exercise will introduce a second input variable.


### Imports

```python
from sklearn.linear_model import SGDRegressor
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

### Generate the data

```python
y = [19, 20, 20.5, 21.5, 22, 23, 23, 25.5, 24, 25]
X = np.array([[0, 1], [1, 1], [2, 3], [3, 2], [4, 3], [4, 4], [5, 4], [6, 5], [7, 5], [7, 6]])
```

### Apply multivariate linear regression

```python
clf = SGDRegressor(max_iter = 1000, penalty = 'none', learning_rate = 'invscaling', eta0 = 0.01)
clf.fit(X, y)
```

### Question 1: What are the coefficients and intercept your model determined

```python
b = clf.intercept_
w = clf.coef_
w1 = w[0]
w2 = w[1]
print "Intercept:", b
print "Coefficients:", w
```

### Question 2: What do the regression coefficients say about the weight of the input variables and the relations of the input variables to the target variable?

First coef: negative value, weight is small

Second coef: positive value, rather high, weight is high

### Plot the data:  As we have 2 input variables and 1 output variable, we'll have to use a 3D plot

```python
x1 = np.array((0,7))
x2 = np.array((0,7))
y3 = eval('x1*w1 + x2*w2 + b')

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],y,X[:,1])
ax.set_xlabel("X1")
ax.set_ylabel("y")
ax.set_zlabel("X2")
ax.plot(x1, y3, x2)
plt.show()
```
